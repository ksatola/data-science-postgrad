# Zadanie
Twoim zadaniem jest wypisanie [Laureatów Pokojowej Nagrody Nobla za rok 2014](http://api.nobelprize.org/v1/prize.json).

```python
response = requests.get("http://api.nobelprize.org/v1/prize.json")
d = json.loads(response.text)
```

```python
d
```

```python
d['prizes'][:2]
```

```python
d['prizes'][0]['laureates'][0]['firstname']
```

```python
years = d['prizes']
for year in years:
    print(4*'-')
    print(year['year'], year['category'])
    if 'laureates' in year.keys():
        laureates = year['laureates']
        for laureate in laureates:
            try:
                print(laureate['firstname'], laureate['surname'])
            except:
                print(laureate['firstname'])
```
